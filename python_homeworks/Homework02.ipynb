**Problem 1.** Using Python, read and parse the data in the given "Counties2010.csv" file opened in the above cell (download from the Canvas HW2 page).
```python
# Problem 1
# Insert your work here

file.readline() # discards header line
Answer1 = []

for line in file:
    county = line.rstrip("\n").split(",")
    Answer1.append(county)
    
for county in Answer1:
    county[2] = int(county[2])
    county[3] = int(county[3])
    county[4] = int(county[4])
    county[5] = float(county[5])
    county[6] = int(county[6])
    county[7] = float(county[7])
    county[8] = float(county[8])

Answer1 = [tuple(county) for county in Answer1]
Answer1
```

**Problem 3.** What is the mean and standard deviation of the education levels?  Compute a  pair (mu, sigma) containing the mean (mu) and standard deviation (sigma) of the "PercentCollegeGrad" attribute over all counties. You are **not** allowed to use library functions for mean and standard deviation.  You must compute them yourself using the following definitions (N = number of tuples). Think about the types of list transforms we discussed in class.

![formula](https://wikimedia.org/api/rest_v1/media/math/render/svg/32e3c0f27c2595926963cc5d8df113e6a12cf917)
https://wikimedia.org/api/rest_v1/media/math/render/svg/32e3c0f27c2595926963cc5d8df113e6a12cf917

Return the result in variable Answer3 as a tuple of the form (mu, sigma).

```python
# Problem 3
# Insert your work here

percent_college_grad = [float(i[5]) for i in Answer1]

def mean(pcd):
    n = len(pcd)
    return ( sum(pcd) / n )

def sd(pcd):
    n = len(pcd)
    mu = mean(pcd)
    return sqrt( (1/n) * sum((x - mu) ** 2 for x in pcd) )



Answer3 = ( mean(percent_college_grad), sd(percent_college_grad) )
Answer3
```
