**Problem 0: (10 points)** For this homework, you will need to first normalize each quantitative column in the above student survey table using z-scores https://en.wikipedia.org/wiki/Standard_score.  You will also find it very helpful to re-order the columns into alphabetical order by column names.  Hint: check out the sort_index() method.

In Answer0, return the z-score table containing only the quantitative columns, with columns names ordered alphabetically, as a Pandas DataFrame.

```python
# Problem 0
# Insert your work here
table.index = table.Name
table = table.iloc[:, 2:]
Answer0 = (table - table.mean()) / table.std()
Answer0.sort_index(axis=1, inplace=True)

Answer0.describe()
```

**Problem 1: (20 points)** Who are the biggest outliers in each quantitative column?  Model the outlier metric as the absolute value of the z-score (i.e. normalized distance from the column mean). For each column, find the person name who is the most outlier in that column, and what their original (survey) value is for that column. For readability, convert the reported values to integers.  Hint: you might find the idxmax() function helpful.

In Answer1, return a DataFrame with 3 columns ['Column', 'Name', 'Value'], with rows sorted alphabetically by 'Column'. 

```python
# Problem 1
# Insert your work here

names = Answer0.abs().idxmax()
columns = Answer0.columns
values = table.loc[names] 
values.sort_index(axis=1, inplace=True)
values = [int(values.iloc[i, i]) for i in range(23)]

df1 = { "Column" : columns,
       "Name" : names,
       "Value" : values }

Answer1 = pd.DataFrame(df1).sort_values(by="Column")

Answer1
```

**Problem 3: (20 points)** What is the multidimensional outlier-ness of each person?  In data analytics, we often try to quantitatively model qualitative concepts that involve multiple dimensions. For example, we may have an intuitive idea of the total "outlier-ness" of a person. Now we want to model the total multi-dimensional outlierness of each person as a single metric that takes into account all the survey questions.

Model the total outlierness of each person as the square root of the sum of squares (SRSS) of that person's whole row of z-scores. (Hint: The axis parameter on pandas reduce operators is handy sometimes.)
![formula](https://wikimedia.org/api/rest_v1/media/math/render/svg/4d2562bd8e6df0c2625fd9c0e0c09ee9b932785d)

In Answer3, return a DataFrame containing two columns ['Name', 'Outlierness'], sorted by Outlierness from most to least.

*Think: Did you land where you expected in the list?*

*Note: You just computed Euclidean distance of each person to the center of mass, also known as the [Mahalanobis distance](https://en.wikipedia.org/wiki/Mahalanobis_distance).*

```python
# Problem 3
# Insert your work here

vals = ((Answer0 - Answer0.mean())**2).sum(axis=1)**0.5

df3 = { "Name": Answer0.index,
        "Outlierness": vals }

Answer3 = pd.DataFrame(df3).sort_values(by="Outlierness", ascending=False)

Answer3
```

**Problem 5: (20 points)** What are the most positively and negatively correlated columns?  In Answer5, return a DataFrame that lists all $p(p-1)/2$ column pairs, sorted by the Pearson correlation values from most positive to most negative, in the form:  \['Column 1', 'Column 2', 'Correlation value'\].  Do not include any pairs twice, do not include column self correlations, and the value in Column 1 should alphabetically precede Column 2.

```python
# Problem 5
# Insert your work here

Answer5 = Answer0.corr()
Answer5 = Answer5.where(np.triu(Answer5).astype(np.bool))
np.fill_diagonal(Answer5.values, np.NaN)
Answer5 = Answer5.stack().reset_index()
Answer5.columns = ['Column1','Column2','Correlation value']
Answer5 = Answer5.sort_values(by="Correlation value", ascending=False).reset_index(drop=True)
Answer5
```
