1. balldontlie API , https://www.balldontlie.io/#introduction
2. Which of these select players had the highest average field goal percentage, three-point percentage, and free throw percentage in the 2019-2020 NBA regular season? 
3. The solution to this question is: Rudy Gobert was had the highest average field goal percentage, Zion Williamson had the highest three-point percentage, and Khris Middleton had the highest free throw percentage in the 2019-2020 NBA regular season. I found the answer by assigning player names to their personal ids and then used the ids to find their respective percentages. Finally, I utilized the max() function to find the highest averages in the three categories and subsequent if statements to match the averages to player names.

```python
import requests
import matplotlib.pyplot as plt

#dictionary of player names and ids
players = {'Giannis Antetokounmpo':0,
                   'Kawhi Leonard':0,
                    'Lebron James':0,
                    'James Harden':0,
                   'Anthony Davis':0,
                     'Alex Caruso':0,
                    'Nikola Jokic':0,
                     'Joel Embiid':0,
                     'Luka Doncic':0,
                  'Damian Lillard':0,
                   'Pascal Siakam':0,
                    'Jimmy Butler':0,
              'Karl-Anthony Towns':0,
                    'Jayson Tatum':0,
               'Russell Westbrook':0,
                     'Paul George':0,
                     'Rudy Gobert':0,
                    'Bradley Beal':0,
                     'Ben Simmons':0,
                 'Khris Middleton':0,
                 'Zion Williamson':0  
}

#adds player ids to player names in players dictionary
for player in players:
    response = requests.get('https://www.balldontlie.io/api/v1/players?search=' + player)
    response = response.json()['data'][0]
    players[player] = response['id']
    
playerStats = {}

#creates a dictionary for each player's percentages.
for player in players:
    response = requests.get('https://www.balldontlie.io/api/v1/season_averages?seasons[]=2019&player_ids[]=' + str(players[player]))
    response = response.json()['data'][0]
    playerStats[player] = [response['fg_pct'], response['fg3_pct'], response['ft_pct']]

#finds max percentages 
max_fg_pct = max([playerStats[i][0] for i in playerStats])
max_fg3_pct = max([playerStats[i][1] for i in playerStats])
max_ft_pct = max([playerStats[i][2] for i in playerStats])

#prints players with max percentages and their names
for player in playerStats:
    if playerStats[player][0] == max_fg_pct:
        print(player, 'had the highest field goal percentage for 2019-2020 NBA season:', str(round(max_fg_pct*100,2)) + '%.\n')
       
    if playerStats[player][1] == max_fg3_pct:
        print(player, 'had the highest three-point field goal percentage for 2019-2020 NBA season:', str(round(max_fg3_pct*100,2)) + '%.\n')
        
    if playerStats[player][2] == max_ft_pct:
        print(player, 'had the highest free throw percentage for 2019-2020 NBA season:', str(round(max_ft_pct*100,2)) + '%.\n')
        
#graphs all player percentages in each category 
playerNames = []
fgPercentages = []
fg3Percentages = []
ftPercentages = []

for guy in playerStats:
    #organizes all players percentages into respective lists 
    fgPercentages.append(playerStats[guy][0]*100)
    fg3Percentages.append(playerStats[guy][1]*100)
    ftPercentages.append(playerStats[guy][2]*100)
    playerNames.append(guy)

fig, ax = plt.subplots()
fig.suptitle('Player Percentages')
ax.plot(playerNames, fgPercentages, label='Field Goal Percentages')
ax.plot(playerNames, fg3Percentages, label='3 Point Field Goal Percentages')
ax.plot(playerNames, ftPercentages,label='Free Throw Percentages')
ax.set_ylabel('Percentages')
ax.set_xticklabels(playerNames, rotation=75, ha='right')
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)
plt.show()
```

1. Cat Facts API, https://alexwohlbruck.github.io/cat-facts/
2. What is the distribution of upvotes on cat facts? What is the most common and the least common upvote the cat facts recieved?
3. The distribution of cat facts is far from normal; It is highly skewed to the right. This suggest a large amount of upvotes were very low, around 1 or 2, and almost no cat facts had high upvote counts. The most common upvote the cat facts recieved was 1, appearing 131 times. The least common upvote the cat facts recieved was 11, with only 1 upvote. 

```python
import requests
import matplotlib.pyplot as plt

#requests cat facts
response = requests.get('https://cat-fact.herokuapp.com/facts')
response = response.json()['all']

#creates dictionary of the different upvotes and their counts
upvotes = {}
for fact in response:
    uvs = fact['upvotes']
    if uvs in upvotes:
        upvotes[uvs] += 1
    else:
        upvotes[uvs] = 1

#finds max and min counts of upvotes
maxUpvoteCount = max([upvotes[i] for i in upvotes])
minUpvoteCount = min([upvotes[i] for i in upvotes])

#filters through the upvotes dictionary to find the key corresponding to the max upvote 
#count and min upvote count, respectively, and creates a list of length 1
maxUpvoteIndex = list(filter(lambda x: upvotes[x] == maxUpvoteCount, upvotes))
maxUpvoteIndex = maxUpvoteIndex[0]
minUpvoteIndex = list(filter(lambda x: upvotes[x] == minUpvoteCount, upvotes))
minUpvoteIndex = minUpvoteIndex[0]

#prints results 
print('The most common upvote the cat facts recieved was', str(maxUpvoteIndex) + ', appearing', str(maxUpvoteCount), 'times.')
print('The least common upvote the cat facts recieved was', str(minUpvoteIndex) + ', appearing only', str(minUpvoteCount), 'time.')

#creates a scatter plot of the data with red points and a green line and includes labels for 
#the max and min counts and their coordinates
votes = [i for i in upvotes]
counts = [upvotes[i] for i in upvotes]
plt.plot(votes,counts,'ro')
plt.plot(votes,counts,'g-')
plt.xlabel('Upvotes')
plt.ylabel('Amount of Upvotes')
plt.title('Distribution of Upvotes')
plt.text(maxUpvoteIndex+.3,maxUpvoteCount-9,'Max Count'+' at:\n    ('+str(maxUpvoteIndex)+', '+str(maxUpvoteCount)+')',fontsize=10,weight='bold')
plt.text(minUpvoteIndex-1.7,minUpvoteCount+9,'Min Count'+' at:\n      ('+str(minUpvoteIndex)+', '+str(minUpvoteCount)+')',fontsize=10,weight='bold')
plt.show()
```

1. CoinDesk API, https://www.coindesk.com/coindesk-api
2. What was the trend of Bitcoin's closing prices for the year 2020 up to now?
3. I used API url links from the beginning of the year until today, November 26th, 2020, and created a plot to show the trend of Bitcoin closing prices. From the plot, a few thing can be infered. First, the stock prices for Bitcoin dropped quickly in March, aligning with the lockdowns due to covid. In addition, the stock price changed very small amounts during the summer months, with rates of change close to zero, as lockdowns continued. As the country began to reopen, prices seem to have increased. Although coronavirus has had a major impact on the economy as a whole, Bitcoin prices seem to have increased over the course of the year up to this point.

```python
import requests 
import matplotlib.pyplot as plt
from pprint import pprint

#dictionary of bitcoin closing price urls for the year 2020 to the present month, November
months = {'Jan':'https://api.coindesk.com/v1/bpi/historical/close.json?start=2020-01-01&end=2020-01-31',
          'Feb':'https://api.coindesk.com/v1/bpi/historical/close.json?start=2020-02-01&end=2020-02-29',
          'Mar':'https://api.coindesk.com/v1/bpi/historical/close.json?start=2020-03-01&end=2020-03-31',
          'Apr':'https://api.coindesk.com/v1/bpi/historical/close.json?start=2020-04-01&end=2020-04-30',
          'May':'https://api.coindesk.com/v1/bpi/historical/close.json?start=2020-05-01&end=2020-05-31',
          'Jun':'https://api.coindesk.com/v1/bpi/historical/close.json?start=2020-06-01&end=2020-06-30',
          'Jul':'https://api.coindesk.com/v1/bpi/historical/close.json?start=2020-07-01&end=2020-07-31',
          'Aug':'https://api.coindesk.com/v1/bpi/historical/close.json?start=2020-08-01&end=2020-08-31',
          'Sep':'https://api.coindesk.com/v1/bpi/historical/close.json?start=2020-09-01&end=2020-09-30',
          'Oct':'https://api.coindesk.com/v1/bpi/historical/close.json?start=2020-10-01&end=2020-10-31',
          'Nov':'https://api.coindesk.com/v1/bpi/historical/close.json?start=2020-11-01&end=2020-11-26',
}

#creates a dictionary of the average closing prices for each month
averages = {}
for month in months:
    url = months[month]
    response = requests.get(url)
    response = response.json()['bpi']
    total = 0
    count = 0
    for date in response:
        total += response[date]
        count += 1 
    average = round((total / count), 2)
    averages[month] = average

#creates a scatter plot of the data with red points and a green line 
monthsList = [i for i in averages]  
avs = [averages[i] for i in averages]
plt.plot(monthsList, avs, 'ro')
plt.plot(monthsList, avs, '-g')
plt.show()
```

1. PunkAPI, https://punkapi.com/documentation/v2 , https://www.beerwulf.com/en-gb/articles-about-craft-beer/beer-figures
2. Which color do most beers have, using European Brewery Convention, or EBC?
3. I first found different ranges of EBCs and their corresponding colors and organized the data into lists. I then found the percentage of beers in each category relative to all of them in the api. I then made a pie chart to visualize the percentages. Based on the chart, amber colored beers are the most common.

```python
import requests 
import matplotlib.pyplot as plt
from pprint import pprint
response = requests.get('https://api.punkapi.com/v2/beers')
response = response.json()


lightBlonde = []
blonde = []
gold = []
amber = []
copper = []
darkCopper = []
darkBrown = []
black = []

#adds names of beers to their respective color category based on their ebc 
for beer in response:
    ebc = beer['ebc']
    if ebc != None:
        if 6 <= ebc <= 9:
            lightBlonde.append(beer['name'])
        elif 9 < ebc <= 12:
            blonde.append(beer['name'])
        elif 12 < ebc <= 20:
            gold.append(beer['name'])
        elif 20 < ebc <= 30:
            amber.append(beer['name'])
        elif 30 < ebc <= 45:
            copper.append(beer['name'])
        elif 45 < ebc <= 75:
            darkCopper.append(beer['name'])
        elif 75 < ebc <= 120:
            darkBrown.append(beer['name'])
        elif ebc > 120:
            black.append(beer['name'])

#gets percentage of each color of beer relative to all the beers in the api
beersCount = len(response)
lightBlondeAve = len(lightBlonde)/beersCount * 100
blondeAve = len(blonde)/beersCount * 100
goldAve = len(gold)/beersCount * 100
amberAve = len(amber)/beersCount * 100
copperAve = len(copper)/beersCount * 100
darkCopperAve = len(darkCopper)/beersCount * 100
darkBrownAve = len(darkBrown)/beersCount * 100
blackAve = len(black)/beersCount * 100

#creates pie chart for the percentages
labels = 'Light Blonde', 'Blonde', 'Gold', 'Amber', 'Copper', 'Dark Copper', 'Dark Brown', 'Black'
sizes = [lightBlondeAve, blondeAve, goldAve, amberAve, copperAve, darkCopperAve, darkBrownAve, blackAve]
fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, autopct='%1.1f%%',shadow=True, startangle=90)
ax.axis('equal')
plt.show()
```
